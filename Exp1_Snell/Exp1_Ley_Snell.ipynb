### Fenómeno


Un haz de luz viaja (línea roja en la fig.) en el aire (_medio 1_) con indice de refracción $n_1$ e incide en un ángulo $\theta_1$ respecto a la normal sobre un medio, en nuestro caso agua (_medio 2_), con indice de refracción $n_2$, se genera por tanto una _refracción_ que medimos con el ángulo $\theta_2$ respecto a la normal.

Sabemos que se puede estudiar bajo **Ley de Snell** como
$$ n_1 \sin \theta_1 = n_2 \sin \theta_2$$

Obtendremos distancias fijas medidas en el esquema experimental dadas por una altura $H$ y una distancia de la base dada por $D$ medidas sin el _medio 2_.

Cuando se provoca el fenómeno al introducir agua con una altura medida $h_i$ o $h$ obtendremos que la proyección a la base del punto donde se incide con los medios será nuestro punto $O$. Denominaremos a las distancias desde $O$ al punto de luz sin medio como $d$, y a donde se proyecta con el medio $d'$ (vease la fig.).

Se requiere determinar la relación entre altura del agua $h$ y la distancia $\ell$ que cambia desde donde el haz de luz incide sin el medio 2 ($d$) y cuando sí incide ($d'$), determinado como
$$ \ell = d - d' $$

Cabe destacar que se puede determinar el indice $n_2$ y para comprobar la teoría bastará fijarse en la diferencia entre los ángulos de refracción que se pueden medir directamente o a través de este esquema.


### Relación $\ell (h)$


Determinamos primero $d$ utilizando semejanza de triangulos
$$ \frac{H}{D} = \frac{h}{d} \quad \Rightarrow \quad d = \frac{H}{D}\, h$$

$$
l- d = d' = h \tan \theta_2 = \frac{n_1 \sin \theta_2}{\cos \theta_2}
$$

Luego, para $d'$, se obtiene de la _ley de Snell_ y relaciones trigonométricas a continuación

$$
\frac{d'}{\sqrt{(d')^2 + h^2}} =  \sin \theta_2 = \frac{n_1 \sin \theta_1}{n_2}
$$

Despejaremos para $d'$, dado el ángulo conocido $\theta_{1i}$ y el indice de refracción del aire según la literatura de $n_1 = 1.00029 \approx 1$.

Se sigue que

$$
\frac{\ell - d}{\sqrt{(\ell - d)^2 + h^2}} = \frac{n_1 \sin \theta_1}{n_2}
$$

Resultando en definitiva

$$
\ell(h) = h \frac{\frac{n_1 \sin \theta_1}{n_2}}{ \sqrt{1 - \frac{n_1^2 \sin^2 \theta_1}{n_2^2}}} + h \left(\frac{D}{H}\right)
$$

$$
\ell(h) =  h \left(\frac{\frac{n_1 \sin \theta_1}{n_2}}{ \sqrt{1 - \frac{n_1^2 \sin^2 \theta_1}{n_2^2}}} + \frac{D}{H}\right)
$$

$$
\ell(h) =  h \left(\frac{D}{H} - \left( \frac{n_2^2}{n_1^2} \frac{1}{\sin^2\theta_1} - 1 \right)^{-1/2} \right)
$$

$$
\ell(h) =  h \left(\frac{D}{H} - \left( \frac{n_2^2}{n_1^2} \left(\frac{H^2}{D^2}+1\right) - 1 \right)^{-1/2} \right)
$$

O también en terminos de $\theta_1$

$$
\ell(h) = h \left( \tan \theta_1 - \left( \frac{n^2_2}{n^2_1}\frac{1}{\sin^2 \theta_1} - 1 \right)^{-1/2} \right)
$$

Lineal para $h$.

Luego, con el análisis gráfico encontraremos la linealización inmediata con pendiente $m$, por lo que obtendremos el **indice de refracción del agua** $n_2$ como

$$n_2 = \sin \theta_1 n_1 \sqrt{\left( \tan \theta_1 - m \right)^{-2}+1}$$
O también
$$n_2 = \frac{n_1 D}{\sqrt{D^2+H^2}}\sqrt{\left(\frac{D}{H}-m\right)^{-2}+1}$$

### Método


Se utilizará un análisis gráfico sobre los resultados de $h$ y $\ell$, teniendo en cuenta la dispersión de este último respecto las mediciones directas $d$ y $d'$. El análisis dará cuenta de la pendiente $m$ donde se determinará la dispersión provocada y demás factores para tener en cuenta en la dispersión con $\theta_1$.

$$\frac{\delta \ell}{\delta h} = \frac{\ell}{h} = m$$

Para cada ángulo se determinará un $n_2$ donde se determinará con estadística simple, la dispersión de estos datos y el mejor valor. Tenga en cuenta que debe utilizarse el estimador de la dispersión, el que tiene $N(N-1)$ en su denominador.

La propagación estará dada por las ecuaciones vistas como

$$\frac{\partial n_2}{\partial \theta_1} = \left(n \cos(\theta_1) \sqrt{1 + \left( -1 + \tan\theta_1x) \right)^2} + \frac{n \sec(\theta_1) \left( -1 + \tan(\theta_1) \right) \tan(\theta_1)}{\sqrt{1 + \left( -1 + \tan(\theta_1) \right)^2}}\right)^2$$
Suficientemente largas como para mejor utilizar la libraría `uncertainties`.

```python
O = 83.9  # (.1) cm
D = 38.0  # 37.6 # (.4) cm
H = 21.9  # (.1) cm

# Incertidumbres
u_D = 0.4  # incertidumbre de D
u_H = 0.1  # incertidumbre de H

# Calcular t
t = np.arctan(D / H)

# Derivadas parciales
partial_t_D = 1 / (1 + (D / H) ** 2) * (1 / H)
partial_t_H = -1 / (1 + (D / H) ** 2) * (D / H**2)

# Propagación de incertidumbre
u_t = np.sqrt((partial_t_D * u_D) ** 2 + (partial_t_H * u_H) ** 2)

# Convertir t y u_t a grados
t_deg = np.degrees(t)
u_t_deg = np.degrees(u_t)

t_deg, u_t_deg
```

```python
# Uso de la libreria uncertainties para la propagacion de incertidumbres
uD = ufloat(38, 0.4)
uH = ufloat(21.9, 0.1)

ut = um.atan(uD / uH)
print(ut)
um.degrees(ut)
```

Note que si se realiza el calculo de esta incertidumbre con la librería `uncertainties` que importamos mas arriba, se obtiene el mismo valor para la incertidumbre que aplicando el calculo directo a traves de la ecuacion correspondiente. Por lo que a partir de este momemento se usara `uncertainties` para la propagacion de incertidumbres.

##### Análisis Gráfico


```python
# Código generado por ChatGPT para una visualización rápida de datos, permitido según directriz del profesor.


# Datos experimentales (pares x, y) ya definidos en el código
# Asegúrate de definir los pares en el formato (x, y)
# Ordenar los valores
ell_values = [1, 2, 2.5, 3.5, 4.5, 5.5, 6.5, 7.3, 8.2, 9.2]

# Incertidumbres
u_ell_values = [0.3, 0.2, 0.3, 0.2, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3]
u_h_values = [0.1] * 10

# Generar los datos en el formato solicitado
datos = [(idx + 1, ell) for idx, ell in enumerate(ell_values)]
u_y = u_ell_values
u_x = u_h_values


# Función para mostrar los datos en formato de tabla
def mostrar_tabla(datos):
    # Convertir los datos a un DataFrame de pandas para visualización en tabla
    df = pd.DataFrame(datos, columns=["X", "Y"])
    print("\nDatos experimentales en forma de tabla:")
    print(df)


# Función para graficar los datos
def graficar_datos(datos, u_x, u_y):
    # Separar los datos en listas de X e Y para graficar
    x_vals, y_vals = zip(*datos)

    # Pendiente "m" y error en la pendiente
    global r
    r = stats.linregress(x_vals, y_vals)
    slope, intercept = r.slope, r.intercept
    slope_err = r.stderr  # Error en la pendiente

    # Crear la gráfica
    plt.figure(figsize=(8, 6))
    plt.errorbar(
        x_vals,
        y_vals,
        xerr=u_x,
        yerr=u_y,
        linestyle="none",
        color="blue",
        capsize=1,
        capthick=1,
        lw=0.4,
        label="Mediciones",
    )
    plt.scatter(
        x_vals, y_vals, marker=".", lw=0.1, color="black", label="Datos experimentales"
    )
    plt.axline(
        xy1=(0, intercept),
        slope=slope,
        color="r",
        linestyle="--",
        label=f"Pendiente: {slope:.2f} ± {slope_err:.2f}",
    )

    # Configurar
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.tight_layout()
    plt.ylabel("Separación $\ell$ [cm]", fontsize=18)
    plt.xlabel("Altura $h$ [cm]", fontsize=18)
    plt.grid(which="major", linewidth=0.2, color="black")
    plt.grid(which="minor", linewidth=0.2)
    plt.minorticks_on()
    plt.grid(True)
    plt.legend()
    plt.show()


# Mostrar los datos en formato de tabla
mostrar_tabla(datos)

# Graficar los datos
graficar_datos(datos, u_x, u_y)
```
